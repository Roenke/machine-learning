# Split the data to train and test

```python
from sklearn import cross_validation
X = data.dropna().drop(['shot_made_flag'], axis=1)
Y = data.dropna()['shot_made_flag']
print(len(X))
print(len(Y))
x_train, x_test, y_train, y_test = cross_validation.train_test_split(X, Y, test_size=0.8, random_state=42)
print(len(x_train))
```

# KNN-model (need nonan values, and normalize features)

```python
def remove_nans(data):
    data = data.dropna(axis=1, how='all')
    data = data.dropna(axis=0, how='any')
    for column_name in data.columns.values:
        try:
            data = data[np.isfinite(data[column_name])]
        except:
            pass
    return data

def process_strings(df):
    for i in range(len(df.columns)):
        if df.dtypes[i] == object:
            column = df.columns[i]
            df[column] = pd.to_numeric(df[column], errors='coerce')
    return df
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier(n_neighbors=3) # neighbors count need to set odd

X = remove_nans(process_strings(X))
clf.fit(X, Y)
res = clf.predict(remove_nans(process_strings(x_test)))
fails_count = 0
for (right, predicted) in zip(res, y_test):
    if right != predicted:
        fails_count += 1
print(fails_count, len(x_test))
```

# Random Forest model

```python
from sklearn import ensemble
model = ensemble.RandomForestClassifier(n_estimators=3)
model.fit(remove_nans(process_strings(x_train)), y_train)
res = model.predict(remove_nans(process_strings(x_test)))
fails_count = 0
for (right, predicted) in zip(res, y_test):
    if right != predicted:
        fails_count += 1
print(fails_count, len(x_test))
```
