```python
from sklearn import datasets
import pandas as pd
from sklearn import preprocessing
boston = datasets.load_boston()
x = pd.DataFrame(preprocessing.scale(boston.data), columns=boston.feature_names)
y = pd.DataFrame(boston.target)
```

```python
from sklearn import cross_validation
fold = cross_validation.KFold(len(x), n_folds=5, shuffle=True, random_state=42)
```

```python
import numpy
from sklearn import neighbors
best_p, max_score = 0., 0.
for p in numpy.linspace(1, 10, 200):
    knn_regressor = neighbors.KNeighborsRegressor(n_neighbors=5, weights='distance', metric='minkowski', p=p)
    res = cross_validation.cross_val_score(knn_regressor, x, y=y, cv=fold, scoring='mean_squared_error')
    print(res)
    score = sum(res) / len(res)
    if score > max_score:
        max_score = score
        best_p = p
(best_p, max_score)
```
