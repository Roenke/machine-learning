```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import *
from sklearn.cross_validation import train_test_split
from sklearn.naive_bayes import MultinomialNB, BernoulliNB
```

```python
data = pd.read_csv('data/PO1.csv')
```

```python
data.head(10)
```

```python
X = data.drop('Decision', axis=1)
X.head(5)

y = data.Decision
y.head(5)
```

```python
for i in X:
    print(set(X[i]))
```

```python
print(set(y))
```

```python
enc = LabelEncoder()
for i in X:
    X[i] = enc.fit_transform(X[i])
y = enc.fit_transform(y)
```

```python
X.head(10)
```

```python
y[1:10]
```

```python
h_enc = OneHotEncoder(sparse=False)
X_enc = h_enc.fit_transform(X)
```

```python
X_enc
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_enc, y, test_size=0.2, random_state=2016)
```

```python
clf = BernoulliNB()
clf = clf.fit(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```
